```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

# Load the pre-trained tokenizer and model
model_name = "joeddav/distilbert-base-uncased-go-emotions-student"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# GoEmotions label list (27 emotions + Neutral)
labels = [
    'admiration', 'amusement', 'anger', 'annoyance', 'approval', 'caring', 'confusion',
    'curiosity', 'desire', 'disappointment', 'disapproval', 'embarrassment', 'excitement',
    'fear', 'gratitude', 'grief', 'joy', 'love', 'nervousness', 'optimism', 'pride',
    'realization', 'relief', 'remorse', 'sadness', 'surprise', 'neutral'
]

# Function to predict emotion
def detect_emotions(text, threshold=0.3):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    with torch.no_grad():
        logits = model(**inputs).logits
        probs = torch.sigmoid(logits)[0]

    # Get emotions above threshold
    results = [(labels[i], float(probs[i])) for i in range(len(labels)) if probs[i] > threshold]

    return sorted(results, key=lambda x: x[1], reverse=True)

# 🔍 Test
text = "I’m really excited and nervous about tomorrow!"
emotions = detect_emotions(text)

print("Detected Emotions:")
for label, score in emotions:
    print(f"{label}: {score:.2f}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from transformers import pipeline
import gradio as gr
import time

# Translation and Summarization Pipelines
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-ar-en")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Article extraction using headless Chrome
def get_arabic_article_text(url):
    try:
        chrome_options = Options()
        chrome_options.add_argument("--headless")
        chrome_options.add_argument("--disable-gpu")
        chrome_options.add_argument("--no-sandbox")
        chrome_options.add_argument('--disable-dev-shm-usage')
        chrome_options.add_argument("user-agent=Mozilla/5.0")

        driver = webdriver.Chrome(options=chrome_options)
        driver.get(url)
        time.sleep(3)  # Allow JS to load

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        driver.quit()

        # Extract all <p> tags (adjust if specific sites use <div> etc.)
        paragraphs = soup.find_all('p')
        text = ' '.join(p.get_text() for p in paragraphs)

        return text.strip() if text else "❌ No content extracted."

    except Exception as e:
        return f"❌ Error extracting article: {e}"

# Main function
def translate_and_summarize(url):
    arabic_text = get_arabic_article_text(url)

    if not arabic_text or arabic_text.startswith("❌"):
        return arabic_text, "", ""

    try:
        translation_output = translator(arabic_text, max_length=1024)
        english_text = translation_output[0]['translation_text']

        summary_output = summarizer(english_text, max_length=130, min_length=30, do_sample=False)
        summary = summary_output[0]['summary_text']

        return arabic_text, english_text, summary

    except Exception as e:
        return arabic_text, "", f"❌ Error during translation/summarization: {e}"

# Gradio UI
demo = gr.Interface(
    fn=translate_and_summarize,
    inputs=gr.Textbox(lines=1, label="📰 Enter Arabic News Article URL"),
    outputs=[
        gr.Textbox(label="📄 Original Arabic Text"),
        gr.Textbox(label="🌍 Translated English Text"),
        gr.Textbox(label="🧠 Summary (English)")
    ],
    title="📰 Arabic News Translator & Summarizer",
    description="Scrapes Arabic news, translates to English, and summarizes it. Now works with CNN Arabic, Al Jazeera, etc.",
    theme="default"
)

# Run the app
demo.launch()
```
